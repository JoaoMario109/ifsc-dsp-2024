## Ex 1

### Signal 1

$$
x_1(n) = 3\delta(n + 2) + 2\delta(n) - \delta(n - 3) + 5\delta(n - 7), \quad -5 < n < 15.
$$

```python
# Implementing: x_1(n) = 3\delta(n + 2) + 2\delta(n) - \delta(n - 3) + 5\delta(n - 7), \quad -5 < n < 15.
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-5, 15)
x = 3 * (n == -2) + 2 * (n == 0) - (n == 3) + 5 * (n == 7)

plt.stem(n, x)
# Adjust grid and axes scales and intervals
plt.grid(alpha = 0.3)
plt.xticks(np.arange(-5, 16, 1))
plt.yticks(np.arange(-2, 7, 1))
# Add labels and title
plt.xlabel('n')
plt.ylabel('x_1(n)')
plt.title('Impulse train')
plt.show()
```

### Signal 2

$$
x_2(n) = \sum_{k=-5}^{5} e^{-|k|} \delta(n - 2k), \quad -10 \leq n \leq 10.
$$

```python
#Implementing x_2(n) = \sum_{k=-5}^{5} e^{-|k|} \delta(n - 2k), \quad -10 \leq n \leq 10.

n = np.arange(-10, 11)
x = np.zeros(n.size)

for k in range(-5, 6):
	x += np.exp(-np.abs(k)) * (n == 2 * k)

plt.stem(n, x)
# Adjust grid and axes scales and intervals
plt.grid(alpha = 0.3)
plt.xticks(np.arange(-10, 11, 1))
plt.yticks(np.arange(0, 1.61, 0.2))
# Add labels and title
plt.xlabel('n')
plt.ylabel('x_2(n)')
plt.title('Impulse train')
```

### Signal 3

$$
\quad x_3(n) = 10u(n) - 5u(n - 5) - 10u(n - 10) + 5u(n - 15). \\
$$

```python
# Implementing \quad x_3(n) = 10u(n) - 5u(n - 5) - 10u(n - 10) + 5u(n - 15)

n = np.arange(-5, 20)
x = 10 * (n >= 0) - 5 * (n >= 5) - 10 * (n >= 10) + 5 * (n >= 15)

plt.stem(n, x)
# Adjust grid and axes scales and intervals
plt.grid(alpha = 0.3)
plt.xticks(np.arange(-5, 21, 1))
plt.yticks(np.arange(-6, 11, 1))
# Add labels and title
plt.xlabel('n')
plt.ylabel('x_3(n)')
plt.title('Impulse train')
```

### Signal 4

$$
\quad x_4(n) = e^{0.1n}[u(n + 20) - u(n - 10)].
$$

```python
# Implementing \quad x_4(n) = e^{0.1n}[u(n + 20) - u(n - 10)]

n = np.arange(-10, 21)
x = np.exp(0.1 * n) * (n >= -20) - np.exp(0.1 * n) * (n >= 10)

plt.stem(n, x)
# Adjust grid and axes scales and intervals
plt.grid(alpha = 0.3)
plt.xticks(np.arange(-10, 21, 1))
plt.yticks(np.arange(-1, 3, 1))
# Add labels and title
plt.xlabel('n')
plt.ylabel('x_4(n)')
plt.title('Impulse train')
```

### Signal 5
