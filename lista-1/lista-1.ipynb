## Ex 1

### Signal 1

$$
x_1(n) = 3\delta(n + 2) + 2\delta(n) - \delta(n - 3) + 5\delta(n - 7), \quad -5 < n < 15.
$$

```python
# Implementing: x_1(n) = 3\delta(n + 2) + 2\delta(n) - \delta(n - 3) + 5\delta(n - 7), \quad -5 < n < 15.

n = np.arange(-5, 15)
x = 3 * Signal.from_impulse(n, -2) + 2 * Signal.from_impulse(n) - Signal.from_impulse(n, 3) + 5 * Signal.from_impulse(n, 7)

x.stem()
```

### Signal 2

$$
x_2(n) = \sum_{k=-5}^{5} e^{-|k|} \delta(n - 2k), \quad -10 \leq n \leq 10.
$$

```python
#Implementing x_2(n) = \sum_{k=-5}^{5} e^{-|k|} \delta(n - 2k), \quad -10 \leq n \leq 10.

n = np.arange(-10, 11)
x = Signal.from_zeros(n)

for k in range(-5, 6):
	x += np.exp(-np.abs(k)) * Signal.from_impulse(n, 2 * k)

x.stem()
```

### Signal 3

$$
\quad x_3(n) = 10u(n) - 5u(n - 5) - 10u(n - 10) + 5u(n - 15). \\
$$

```python
# Implementing \quad x_3(n) = 10u(n) - 5u(n - 5) - 10u(n - 10) + 5u(n - 15)

n = np.arange(-5, 20)
x = 10 * Signal.from_step(n) - 5 * Signal.from_step(n, 5) - 10 * Signal.from_step(n, 10) + 5 * Signal.from_step(n, 15)

x.stem()
```

### Signal 4

$$
\quad x_4(n) = e^{0.1n}[u(n + 20) - u(n - 10)].
$$

```python
# Implementing \quad x_4(n) = e^{0.1n}[u(n + 20) - u(n - 10)]

n = np.arange(-10, 21)
x = Signal.from_exponential(1, 0.1, n) * (Signal.from_step(n, -20) - Signal.from_step(n, 10))

x.stem()
```

### Signal 5
$$
\quad x_5(n) = 5[\cos(0.49\pi n) + \cos(0.51\pi n)], \quad -200 \leq n \leq 200. \text{ Comment on the waveform shape.}
$$

```python
# Implementing \quad x_5(n) = 5[\cos(0.49\pi n) + \cos(0.51\pi n)], \quad -200 \leq n \leq 200

n = np.arange(-200, 201)
x = 5 * (Signal.from_cosine(0.49 * np.pi, n) + Signal.from_cosine(0.51 * np.pi, n))

x.stem()
```

### Signal 6
$$
\quad x_6(n) = 2 \sin(0.01\pi n) \cos(0.5\pi n), \quad -200 \leq n \leq 200. \text{ Comment on the waveform shape.}
$$

```python
# Implementing \quad x_6(n) = 2 \sin(0.01\pi n) \cos(0.5\pi n), \quad -200 \leq n \leq 200

n = np.arange(-200, 201)
x = 2 * Signal.from_sine(0.01 * np.pi, n) * Signal.from_cosine(0.5 * np.pi, n)

x.stem()
```

### Signal 7
$$
\quad x_7(n) = e^{-0.05n} \sin(0.1\pi n + \pi/3), \quad 0 \leq n \leq 100. \text{ Comment on the waveform shape.}
$$

```python
# Implementing \quad x_7(n) = e^{-0.05n} \sin(0.1\pi n + \pi/3), \quad 0 \leq n \leq 100

n = np.arange(0, 101)
x = Signal.from_exponential(-0.05, n) * Signal.from_sine(0.1 * np.pi, n, np.pi/3)

x.stem()
```

### Signal 8
$$
\quad x_8(n) = e^{0.01n} \sin(0.1\pi n), \quad 0 \leq n \leq 100. \text{ Comment on the waveform shape.}
$$

```python
# Implementing \quad x_8(n) = e^{0.01n} \sin(0.1\pi n), \quad 0 \leq n \leq 100

n = np.arange(0, 101)
x = Signal.from_exponential(0.01, n) * Signal.from_sine(0.1 * np.pi, n)

x.stem()
```

### Signal 1

$$
\tilde{x}_1(n) = \{ \ldots, -2, -1, 0, 1, 2, \ldots \}_{\text{periodic}} \quad \text{Plot 5 periods.} \\
$$

```python
# Implementing \tilde{x}_1(n) = \{ \ldots, -2, -1, 0, 1, 2, \ldots \}_{\text{periodic}} \quad \text{Plot 5 periods.}
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0, 25)
x = np.array([-2, -1, 0, 1, 2])
y = np.tile(x, 5)

plt.stem(n, y)
# Adjust grid and axes scales and intervals
plt.grid(alpha = 0.3)
plt.xticks(np.arange(0, 126, 25))
plt.yticks(np.arange(-2, 3, 1))
# Add labels and title
plt.xlabel('n')
plt.ylabel('x_1(n)')
plt.title('Impulse train')
```
